```python
import pandas as pd

# Replace with the correct full path if needed
file_path = "/Users/monish/Downloads/historical_automobile_sales.csv"
df = pd.read_csv(file_path)

print("Data downloaded and read into a dataframe!")
```

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

yearly_sales.plot(x='Year', y='Automobile_Sales', kind='line', marker='o', title='Yearly Automobile Sales')
```

```python
df_recession = df[df['Recession'] == True]
df_recession['Year'] = pd.to_datetime(df_recession['Date']).dt.year
vehicle_sales_recession = df_recession.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack()

vehicle_sales_recession.plot(title='Vehicle Sales Trends During Recession')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['Year'] = pd.to_datetime(df['Date']).dt.year
sns.lineplot(data=df, x='Year', y='Automobile_Sales', hue='Vehicle_Type', style='Recession')
plt.title('Sales Trend by Vehicle Type: Recession vs Non-Recession')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

recession_data = df[df['Recession'] == True].groupby('Year')['GDP'].mean()
non_recession_data = df[df['Recession'] == False].groupby('Year')['GDP'].mean()

recession_data.plot(ax=axes[0], title='GDP During Recession')
non_recession_data.plot(ax=axes[1], title='GDP During Non-Recession')
plt.tight_layout()
```

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month
monthly_sales = df.groupby('Month')['Automobile_Sales'].sum().reset_index()
plt.scatter(monthly_sales['Month'], monthly_sales['Automobile_Sales'],
            s=monthly_sales['Automobile_Sales'] / 100, alpha=0.6)
plt.title('Impact of Seasonality on Sales')
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.show()
```

```python
import pandas as pd
import dash
from dash import dcc, html, Input, Output
import plotly.express as px

# Load the data
df = pd.read_csv("historical_automobile_sales.csv")

# Create the Dash app (Task 2.1)
app = dash.Dash(__name__)
app.title = "Automobile Sales Dashboard"

# Unique vehicle types and years for dropdown options (Task 2.2)
vehicle_types = df['Vehicle_Type'].unique()
years = sorted(df['Year'].unique())

# App layout (Tasks 2.2 and 2.3)
app.layout = html.Div([
    # Title (Task 2.1)
    html.H1("Automobile Sales Dashboard", style={'textAlign': 'center', 'color': '#503D36', 'font-size': 24}),

    # Dropdowns (Task 2.2)
    html.Div([
        html.Label("Select Vehicle Type:", style={'margin-right': '10px'}),
        dcc.Dropdown(
            id='vehicle-type-dropdown',
            options=[{'label': vt, 'value': vt} for vt in vehicle_types],
            value=vehicle_types[0],
            placeholder="Select a Vehicle Type",
            style={'width': '80%', 'padding': '3px', 'font-size': '20px', 'text-align-last': 'center'}
        ),
        html.Br(),
        html.Label("Select Year:", style={'margin-right': '10px'}),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': str(y), 'value': y} for y in years],
            value=years[0],
            placeholder="Select a Year",
            style={'width': '80%', 'padding': '3px', 'font-size': '20px', 'text-align-last': 'center'}
        ),
    ], style={'display': 'flex', 'flex-direction': 'column', 'align-items': 'center', 'margin-top': '20px'}), # Centered dropdowns

    html.Br(),

    # Output display division (Task 2.3)
    html.Div(
        id='output-container',
        className='output-div',
        style={'textAlign': 'center', 'fontSize': 18, 'marginTop': '20px', 'marginBottom': '20px', 'color': '#503D36'}
    ),

    # Graphs for recession and yearly reports (Tasks 2.5 & 2.6)
    html.Div([
        dcc.Graph(id='recession-graph', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='yearly-graph', style={'width': '48%', 'display': 'inline-block'})
    ], style={'display': 'flex', 'justify-content': 'space-around'}) # Arranged graphs side-by-side
])

# Callback for updating output and graphs (Tasks 2.4, 2.5, 2.6)
@app.callback(
    [Output('output-container', 'children'),
     Output('recession-graph', 'figure'),
     Output('yearly-graph', 'figure')],
    [Input('vehicle-type-dropdown', 'value'),
     Input('year-dropdown', 'value')]
)
def update_dashboard(selected_vehicle, selected_year):
    # Filter for selected vehicle type
    filtered_df = df[df['Vehicle_Type'] == selected_vehicle]

    # Recession period data for selected vehicle (Task 2.5)
    recession_df = filtered_df[filtered_df['Recession'] == True]
    if not recession_df.empty:
        recession_fig = px.pie(
            recession_df,
            names='Month',
            values='Automobile_Sales',
            title=f'Monthly Average Sales for {selected_vehicle} (Recession)',
            hole=0.4,
            color_discrete_sequence=px.colors.sequential.RdBu
        )
    else:
        recession_fig = px.pie(title="No Recession Data Available for Selected Vehicle")

    # Yearly data for selected year and vehicle (Task 2.6)
    year_df = filtered_df[filtered_df['Year'] == selected_year]
    if not year_df.empty:
        yearly_fig = px.bar(
            year_df,
            x='Month',
            y='Automobile_Sales',
            title=f'Monthly Sales for {selected_vehicle} in {selected_year}',
            color='Automobile_Sales', # Color bars based on sales value
            color_continuous_scale=px.colors.sequential.Viridis
        )
    else:
        yearly_fig = px.bar(title="No Data Available for Selected Year")

    # Output text update (Task 2.4)
    output_text = f"Showing data for Vehicle Type: {selected_vehicle}, Year: {selected_year}"

    return output_text, recession_fig, yearly_fig

# Run the app correctly (updated method)
if __name__ == '__main__':
    app.run(debug=True)
```
